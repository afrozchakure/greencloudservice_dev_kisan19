```python
def get_dataframe(table_name):
    try:
        connection = mysql.connector.connect(host = 'localhost',
                                            database = 'greencloudservice_dev_kisan19',
                                            user = 'root',
                                            password = 'afroz123')
        mycursor = connection.cursor()
        mysql_select_query = "SELECT * FROM " + str(table_name)
        mycursor.execute(mysql_select_query)

        rows = mycursor.fetchall()  # Fetches all rows with following entries

        mycursor.execute("SHOW columns FROM " + str(table_name))
        x = [column[0] for column in mycursor.fetchall()]

        # print(myresult) # prints the whole list of results

        print("Record displayed sucessfully in Laptop")
        mycursor.close()
        return pd.DataFrame( [[ij for ij in i] for i in rows], columns= [x for x in x])
        

    except mysql.connector.Error as error:
        print("Failed to display the record from analytics table {}".format(error))

    finally:
        if(connection.is_connected()):
            connection.close()
            print("MySQL connection is closed")
```
